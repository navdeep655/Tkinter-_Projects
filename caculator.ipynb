### Description: This is a basic calculator implemented using Python's Tkinter library. While using sin,cos,log,sqrt use parenthisis() 

```python
# Import the Tkinter library for GUI
import tkinter as tk
 # Import math module for mathematical functions
import math
 #Handles button click events.
def button_click(value):
    current=entry.get()
    if value=="=":  # Get the current text from the entry field
        result=eval(current)
        entry.delete(0,tk.END)
        entry.insert(tk.END,result)
    elif value=="X": # Remove last digit one by one
        if len(current) > 0:
            entry.delete(len(current) - 1, tk.END)
    elif value=="C": # Clear all input
       entry.delete(0,tk.END)
    else:  # Insert the clicked button's value into the entry
        entry.insert(tk.END,value)
        
# Initialize main Tkinter window 
root=tk.Tk()
root.title("calculator")
root.configure(bg="grey")

# Entry widget for input and output
entry=tk.Entry(root,width=30,borderwidth=2,font=("Arial", 13),justify="right",relief="solid")
entry.grid(row=0,column=0,columnspan=6,padx=5,pady=5)

# Define calculator buttons and place them in the grid
button_7=tk.Button(root,text="7",width=4,height=2,command=lambda:button_click("7"))
button_7.grid(row=1,column=1,padx=2,pady=2)
button_8=tk.Button(root,text="8",width=4,height=2,command=lambda:button_click("8"))
button_8.grid(row=1,column=2,padx=2,pady=2)
button_9=tk.Button(root,text="9",width=4,height=2,command=lambda:button_click("9"))
button_9.grid(row=1,column=3,padx=2,pady=2)
button_add=tk.Button(root,text="+",width=4,height=2,command=lambda:button_click("+"))
button_add.grid(row=1,column=4,padx=2,pady=2)
button_sub=tk.Button(root,text="-",width=4,height=2,command=lambda:button_click("-"))
button_sub.grid(row=1,column=5,padx=2,pady=2)

button_4=tk.Button(root,text="4",width=4,height=2,command=lambda:button_click("4"))
button_4.grid(row=2,column=1,padx=2,pady=2)
button_5=tk.Button(root,text="5",width=4,height=2,command=lambda:button_click("5"))
button_5.grid(row=2,column=2,padx=2,pady=2)
button_6=tk.Button(root,text="6",width=4,height=2,command=lambda:button_click("6"))
button_6.grid(row=2,column=3,padx=2,pady=2)
button_prod=tk.Button(root,text="*",width=4,height=2,command=lambda:button_click("*"))
button_prod.grid(row=2,column=4,padx=2,pady=2)
button_div=tk.Button(root,text="/",width=4,height=2,command=lambda:button_click("/"))
button_div.grid(row=2,column=5,padx=2,pady=2)

button_1=tk.Button(root,text="1",width=4,height=2,command=lambda:button_click("1"))
button_1.grid(row=3,column=1,padx=2,pady=2)
button_2=tk.Button(root,text="2",width=4,height=2,command=lambda:button_click("2"))
button_2.grid(row=3,column=2,padx=2,pady=2)
button_3=tk.Button(root,text="3",width=4,height=2,command=lambda:button_click("3"))
button_3.grid(row=3,column=3,padx=2,pady=2)
button_log=tk.Button(root,text="Log",width=4,height=2,command=lambda:button_click("math.log10"))
button_log.grid(row=3,column=4,padx=2,pady=2)
button_sq=tk.Button(root,text="âˆš",width=4,height=2,command=lambda:button_click("math.sqrt"))
button_sq.grid(row=3,column=5,padx=2,pady=2)

button_0=tk.Button(root,text="0",width=4,height=2,command=lambda:button_click("0"))
button_0.grid(row=4,column=3,padx=2,pady=2)
button_eq=tk.Button(root,text="=",width=4,height=2,command=lambda:button_click("="),font=("Arial", 10),bg="green",fg="white")
button_eq.grid(row=4,column=2,padx=2,pady=2)
button_c=tk.Button(root,text="C",width=4,height=2,command=lambda:button_click("C"),font=("Arial", 10,"bold"),bg="orange",fg="white")
button_c.grid(row=4,column=1,padx=2,pady=2)
button_sin=tk.Button(root,text="Sin",width=4,height=2,command=lambda:button_click("math.sin"))
button_sin.grid(row=4,column=4,padx=2,pady=2)
button_cos=tk.Button(root,text="Cos",width=4,height=2,command=lambda:button_click("math.cos"))
button_cos.grid(row=4,column=5,padx=2,pady=2)
button_close_paren = tk.Button(root, text=")", width=4, height=2, command=lambda: button_click(")"))
button_close_paren.grid(row=2, column=6, padx=2, pady=2)
button_close_paren = tk.Button(root, text="(", width=4, height=2, command=lambda: button_click("("))
button_close_paren.grid(row=3, column=6, padx=2, pady=2)
button_close_paren = tk.Button(root, text="X ", width=4, height=1, command=lambda: button_click("X"),fg="white",bg="red")
button_close_paren.grid(row=0, column=6, padx=2, pady=2)
button_close_paren = tk.Button(root, text=".", width=4, height=2, command=lambda: button_click("."),font=("Arial",10,"bold"))
button_close_paren.grid(row=4, column=6, padx=2, pady=2)
button_close_paren = tk.Button(root, text="%", width=4, height=2, command=lambda: button_click("%"))
button_close_paren.grid(row=1, column=6, padx=2, pady=2)

# Run the Tkinter event loop
root.mainloop()
```
